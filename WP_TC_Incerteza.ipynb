## Cálculos de médias e retornos

```python
#df_mean['IVOL_BR_SMA21'] = df_mean['IVOL_BR'].rolling(21).mean()
```

```python
df_mean['TC Incert_SMA21'] = df_mean['TC Incert'].rolling(21).mean()
df_mean['TC Incert_SMA50'] = df_mean['TC Incert'].rolling(50).mean()
df_mean['TC Incert_SMA100'] = df_mean['TC Incert'].rolling(100).mean()
df_mean['TC Incert_SMA200'] = df_mean['TC Incert'].rolling(200).mean()
```

```python
df_mean['TC Expect_SMA21'] = df_mean['TC Incert Expect'].rolling(21).mean()
df_mean['TC Expect_SMA50'] = df_mean['TC Incert Expect'].rolling(50).mean()
df_mean['TC Expect_SMA100'] = df_mean['TC Incert Expect'].rolling(100).mean()
df_mean['TC Expect_SMA200'] = df_mean['TC Incert Expect'].rolling(200).mean()
```

```python
df_mean["Incert<SMA21"] = np.where((df_mean['TC Incert'].shift(1)) < (df_mean['TC Incert_SMA21'].shift(1)), 1, 0)
df_mean["Incert<SMA50"] = np.where((df_mean['TC Incert'].shift(1)) < (df_mean['TC Incert_SMA50'].shift(1)), 1, 0)

df_mean["Incert Baixa_SMA21"] = df_mean["TC Incert"] * df_mean["Incert<SMA21"]
df_mean["Incert Baixa_SMA50"] = df_mean["TC Incert"] * df_mean["Incert<SMA50"]
```

```python
df_mean["Incert>SMA21"] = np.where((df_mean['TC Incert'].shift(1)) >= (df_mean['TC Incert_SMA21'].shift(1)), 1, 0)
df_mean["Incert>SMA50"] = np.where((df_mean['TC Incert'].shift(1)) >= (df_mean['TC Incert_SMA50'].shift(1)), 1, 0)

df_mean["Incert Alta_SMA21"] = df_mean["TC Incert"] * df_mean["Incert>SMA21"]
df_mean["Incert Alta_SMA50"] = df_mean["TC Incert"] * df_mean["Incert>SMA50"]
```

```python
df_mean["Incert Expect<SMA21"] = np.where((df_mean['TC Incert Expect'].shift(1)) < (df_mean['TC Expect_SMA21'].shift(1)), 1, 0)
df_mean["Incert Expect<SMA50"] = np.where((df_mean['TC Incert Expect'].shift(1)) < (df_mean['TC Expect_SMA50'].shift(1)), 1, 0)

df_mean["TC Expect Baixa_SMA21"] = df_mean["TC Incert Expect"] * df_mean["Incert Expect<SMA21"]
df_mean["TC Expect Baixa_SMA50"] = df_mean["TC Incert Expect"] * df_mean["Incert Expect<SMA50"]
```

```python
df_mean["Incert Expect>SMA21"] = np.where((df_mean['TC Incert Expect'].shift(1)) >= (df_mean['TC Expect_SMA21'].shift(1)), 1, 0)
df_mean["Incert Expect>SMA50"] = np.where((df_mean['TC Incert Expect'].shift(1)) >= (df_mean['TC Expect_SMA50'].shift(1)), 1, 0)

df_mean["TC Expect Alta_SMA21"] = df_mean["TC Incert Expect"] * df_mean["Incert Expect>SMA21"]
df_mean["TC Expect Alta_SMA50"] = df_mean["TC Incert Expect"] * df_mean["Incert Expect>SMA50"]
```

```python
df_mean['IBOV_Incert_Baixa'] = ((df_mean['Incert<SMA21'] * df_mean['IBOV_var']) + (df_mean['Incert>SMA21'] * df_mean['LFT_var'])) - 0.0000476 # 0,10% de custo operacional mensal
df_mean['IBOV_Incert_Alta'] = df_mean['Incert>SMA21'] * ((df['IBOV_var']*(-1)) + df_mean['LFT_var']) - 0.0000476 # 0,10% de custo operacional mensal
df_mean['IBOV_Incert_SMA21'] = df_mean['IBOV_Incert_Alta'] + df_mean['IBOV_Incert_Baixa']

df_mean['IBOV_Incert>SMA21_cum'] = (1+df_mean['IBOV_Incert_Alta']).cumprod()-1
df_mean['IBOV_Incert<SMA21_cum'] = (1+df_mean['IBOV_Incert_Baixa']).cumprod()-1
df_mean['IBOV_Incert_SMA21_cum'] = (1+df_mean['IBOV_Incert_SMA21']).cumprod()-1
```

```python
df_mean['IBOV_Expect_Baixa'] = (df_mean['Incert Expect<SMA21'] * df_mean['IBOV_var']) + (df_mean['Incert Expect>SMA21'] * df_mean['LFT_var']) - 0.0000476 # 0,10% de custo operacional mensal
df_mean['IBOV_Expect_Alta'] = df_mean['Incert Expect>SMA21'] * ((df['IBOV_var']*(-1)) + df_mean['LFT_var']) - 0.0000476 # 0,10% de custo operacional mensal
df_mean['IBOV_Expect_SMA21'] = df_mean['IBOV_Expect_Alta'] + df_mean['IBOV_Expect_Baixa']

df_mean['IBOV_Expect>SMA21_cum'] = (1+df_mean['IBOV_Expect_Alta']).cumprod()-1
df_mean['IBOV_Expect<SMA21_cum'] = (1+df_mean['IBOV_Expect_Baixa']).cumprod()-1
df_mean['IBOV_Expect_SMA21_cum'] = (1+df_mean['IBOV_Expect_SMA21']).cumprod()-1
```

```python
df_mean['IBOV_cum'] = (1+df_mean['IBOV_var']).cumprod()-1
```

```python
df_mean['TC<Expect_var'] = df_mean['TC Incert Expect_var'] * df_mean['Incert Expect<SMA21']
```

## Cálculos das estatísticas das carteiras

```python
# Retorno anualizado

df_mean['IBOV_Ret'] = (1+df_mean['IBOV_var']).rolling(window=252).agg(lambda x : x.prod())-1
df_mean['IBOV_Incert_Ret'] = (1+df_mean['IBOV_Incert_SMA21']).rolling(window=252).agg(lambda x : x.prod())-1
df_mean['IBOV_Expect_Ret'] = (1+df_mean['IBOV_Expect_SMA21']).rolling(window=252).agg(lambda x : x.prod())-1
df_mean['IBOV<Expect_Ret'] = (1+df_mean['IBOV_Expect_Baixa']).rolling(window=252).agg(lambda x : x.prod())-1

print(df_mean[['IBOV_Ret', 'IBOV_Incert_Ret', 'IBOV_Expect_Ret', 'IBOV<Expect_Ret']].head(255))
```

```python
# Volatilidade

df_mean['IBOV_Vol'] = df_mean['IBOV_var'].rolling(window=252).std() * np.sqrt(252)
df_mean['IBOV_Incert_Vol'] = df_mean['IBOV_Incert_SMA21'].rolling(window=252).std() * np.sqrt(252)
df_mean['IBOV_Expect_Vol'] = df_mean['IBOV_Expect_SMA21'].rolling(window=252).std() * np.sqrt(252)
df_mean['IBOV<Expect_Vol'] = df_mean['IBOV_Expect_Baixa'].rolling(window=252).std() * np.sqrt(252)

print(df_mean[['IBOV_Vol', 'IBOV_Incert_Vol', 'IBOV_Expect_Vol', 'IBOV<Expect_Vol']].tail(5))
```

```python
# Tracking Error

df_mean['IBOV_TE'] = df_mean['IBOV_Vol'] * (df_mean['IBOV_Ret'] - df_mean['IBOV_Ret'])
df_mean['IBOV_Incert_TE'] = df_mean['IBOV_Incert_Vol'] * (df_mean['IBOV_Incert_Ret'] - df_mean['IBOV_Ret'])
df_mean['IBOV_Expect_TE'] = df_mean['IBOV_Expect_Vol'] * (df_mean['IBOV_Expect_Ret'] - df_mean['IBOV_Ret'])
df_mean['IBOV<Expect_TE'] = df_mean['IBOV<Expect_Vol'] * (df_mean['IBOV<Expect_Ret'] - df_mean['IBOV_Ret'])

print(df_mean[['IBOV_TE', 'IBOV_Incert_TE', 'IBOV_Expect_TE', 'IBOV<Expect_TE']].tail(5))
```

```python
# Desvio-padrão anualizado

df_mean['IBOV_DP'] = df_mean['IBOV_Ret'].rolling(window=252).std()
df_mean['IBOV_Incert_DP'] = df_mean['IBOV_Incert_Ret'].rolling(window=252).std()
df_mean['IBOV_Expect_DP'] = df_mean['IBOV_Expect_Ret'].rolling(window=252).std()
df_mean['IBOV<Expect_DP'] = df_mean['IBOV<Expect_Ret'].rolling(window=252).std()

print(df_mean[['IBOV_DP', 'IBOV_Incert_DP', 'IBOV_Expect_DP', 'IBOV<Expect_DP']].tail(5))
```

```python
# Sharpe Ratio

df_mean['LFT_Ret'] = (1+df_mean['LFT_var']).rolling(window=252).agg(lambda x : x.prod())-1

df_mean['IBOV_SR'] = (df_mean['IBOV_Ret'] - df_mean['LFT_Ret']) / df_mean['IBOV_DP']
df_mean['IBOV_Incert_SR'] = (df_mean['IBOV_Incert_Ret'] - df_mean['LFT_Ret']) / df_mean['IBOV_Incert_DP']
df_mean['IBOV_Expect_SR'] = (df_mean['IBOV_Expect_Ret'] - df_mean['LFT_Ret']) / df_mean['IBOV_Expect_DP']
df_mean['IBOV<Expect_SR'] = (df_mean['IBOV<Expect_Ret'] - df_mean['LFT_Ret']) / df_mean['IBOV<Expect_DP']

print(df_mean[['IBOV_SR', 'IBOV_Incert_SR', 'IBOV_Expect_SR', 'IBOV<Expect_SR']].tail(5))
```

## Drawdown

```python
# Compute the wealth index by starting with 100 dollars

df_10y['IBOV_wealth_index'] = 100 * (1+df_10y['IBOV_var']).cumprod() # The starting value won't matter with drawdowns
df_10y['IBOV_previous_peaks'] = df_10y['IBOV_wealth_index'].cummax() # Compute the previous peaks 
df_10y['IBOV_Drawdown'] = (df_10y['IBOV_wealth_index'] - df_10y['IBOV_previous_peaks'])/df_10y['IBOV_previous_peaks'] # Calculate the drawdown in percentage

print(df_10y[['IBOV_wealth_index', 'IBOV_previous_peaks', 'IBOV_Drawdown']].head())
```

```python
# Compute the wealth index by starting with 100 dollars

df_10y['Expect_wealth_index'] = 100 * (1+df_10y['IBOV_Expect_SMA21']).cumprod() # The starting value won't matter with drawdowns
df_10y['Expect_previous_peaks'] = df_10y['Expect_wealth_index'].cummax() # Compute the previous peaks 
df_10y['Expect_Drawdown'] = (df_10y['Expect_wealth_index'] - df_10y['Expect_previous_peaks'])/df_10y['Expect_previous_peaks'] # Calculate the drawdown in percentage

print(df_10y[['Expect_wealth_index', 'Expect_previous_peaks', 'Expect_Drawdown']].head())
```

```python
# Check for Stationarity

from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    for key,val in result[4].items():
        out[f'critical value ({key})']=val
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```
